```python
import pandas as pd
data=pd.read_csv("C:/Users/Dell/Downloads/Bengaluru Ola.csv")
print("Show first and last 10 data")
print(data.head(10))
print(data.head(10))
#to check the data types
print("Data Types:\n")
print(data.info())

#Check if there are missing values
print("\nMisiing values in each column")
print(data.isnull().sum())

#Handle missing values
data['Reason for Cancelling by Customer'].fillna('Not Applicable',inplace=True)
data['Reason for Cancelling by Driver'].fillna('Not Applicable',inplace=True)
data['Incomplete Rides Reason'].fillna('Not Applicable',inplace=True)
data['Payment Method'].fillna('Unknown',inplace=True)
data['Booking Value'].fillna(data['Booking Value'].median(), inplace=True)
data['Ride Distance'].fillna(data['Ride Distance'].median(), inplace=True)
data['Driver Ratings'].fillna(data['Driver Ratings'].mean(), inplace=True)
data['Customer Rating'].fillna(data['Customer Rating'].mean(), inplace=True)
data.drop(columns=['Avg VTAT'],inplace=True)
data.drop(columns=['Avg CTAT'],inplace=True)


print("\nMissing values after cleaning:")
print(data.isnull().sum())

#step 9: save the data into csv file
.to_csv('ola_cleaned.csv',index=False)
print('\n cleaned dataset save to csv file')
```
